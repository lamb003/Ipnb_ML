```python
# import the necessary packages
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
#read the data
data = np.array(pd.read_csv("data/simple.txt", header = None))
Y = data[:,-1]
X = data[:,:-1]
```

The logistic regression Hypothesis is defined by: 

$$h_{\theta}(x) = g(\theta^{T}x)$$

Where g is the sigmoid function or perceptron:

$$g(z) = \dfrac{1}{1 + e^{-z}}$$




```python
clf = linear_model.LogisticRegression()
clf = clf.fit(X,Y)
```

```python
#Visualize the data
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

# fig = plt.figure()
# ax = fig.add_subplot(111)
f, ax = plt.subplots(figsize=(8, 6))
for i in xrange(len(Y)):
    if Y[i] == 0:
        ax.plot(X[i,0], X[i,1], "x", color = 'b' )
    elif Y[i] == 1:
        ax.plot(X[i,0], X[i,1], "o", color = 'r' )


ax.set_xlabel('X1')
ax.set_ylabel('X2')


# Create a mesh grid to depict the probability "bands" for the classification
xx, yy = np.mgrid[20:110:1, 10:120:1]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = clf.predict_proba(grid)[:, 1].reshape(xx.shape)
contour = ax.contourf(xx, yy, probs, 25, cmap="RdBu",
                      vmin=0, vmax=1)
ax_c = f.colorbar(contour)
ax_c.set_label("$P(y = 1)$")
ax_c.set_ticks([0, .25, .5, .75, 1])

ax.scatter(X[100:,0], X[100:, 1], c=Y[100:], s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)

ax.set(aspect="equal",
       xlim=(10, 120), ylim=(10, 120),
       xlabel="$X_1$", ylabel="$X_2$")

plt.show()
```

####The figure above shows the decision bondary and the associated probabilities associated with the point's label
